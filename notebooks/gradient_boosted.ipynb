```python
import os
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_digits
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.multioutput import MultiOutputClassifier

```

```python
df = pd.read_csv("../data/extracted_features_df.csv")
```

```python
df.head()
```

```python
df = df.rename(columns = {"patient_id_x":"patient_id"})
```

```python
#Binary encoding data
non_numerics = df.select_dtypes(include = ['object'])
non_numerics
```

```python
#Checking the number of columns and rows

print(f'The number of rows a in this dataframe are {df.shape[0]} and number of columns is {df.shape[1]}')
```

```python
#Replacing alphabetic categorical data with numeric binary values
'''
Lives - U(Urban) = 1
      - R (Rural) = 0
Gender - M(Male) = 1
        -F(Female) = 0
'''


#Lives
df['Lives'].replace('U',1,inplace=True)
df['Lives'].replace('R',0,inplace= True)
#Gender
df['Gender'].replace('M',1,inplace=True)
df['Gender'].replace('F',0,inplace= True)
df
```

```python
mfcc_deviation_df = df["mfcc devation"].apply(pd.Series)
mfcc_deviation_df.columns = [f"mfcc_dev_{i}" for i in mfcc_deviation_df.columns]
df = df.drop(columns=["mfcc devation"]).join(mfcc_deviation_df)
df = df.drop(columns = "mfcc_dev_0")
```

```python
#dropping other non-numeric values
df.drop([ "file_key","audio_filename_base","patient_id"],axis=1,inplace=True)
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python

print(f'The number of non-empty rows a in this dataframe are {df.shape[0]} and number of columns is {df.shape[1]}.')
```

```python
#labels
y = df[['AS', 'AR',
       'MR', 'MS', 'N']]
#features
X = df.drop(['AS','MR',"AR","MS","N"],axis = 'columns')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1234,test_size = 0.3)
X_train.to_csv('X_train.csv')
X_test.to_csv('X_test.csv')
y_train.to_csv('y_train.csv')
y_test.to_csv('y_test.csv')
```

```python
gbdt_model = GradientBoostingClassifier(n_estimators=300,
                                 learning_rate=0.05,
                                 random_state=100,
                                 max_features=5 )

gbdt_model
```

```python
multilabel_gbdt_model = MultiOutputClassifier(estimator=gbdt_model)
multilabel_gbdt_model
```

```python
multilabel_gbdt_model.fit(X_train, y_train)

y_gbdt_pred = multilabel_gbdt_model.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_gbdt_pred)
print("Gradient Boosting Classifier accuracy is : {:.2f}".format(acc))
```

```python
#hyperparameter tuning FOR Gradient Boosted Decision Tree
print('Running Grid Search...')

param_grid = {
    'estimator__n_estimators': [50, 100, 200],
    'estimator__max_depth': [3, 5, 7],
    'estimator__learning_rate': [0.01, 0.1, 0.2]
}

grid = GridSearchCV(estimator=gbdt_model, 
                    param_grid=param_grid, 
                    cv=9,
                    verbose = 2)

grid_search = grid.fit(X_train,y_train)
print('Done')
```

```python
best_params = grid_search.best_params_
gbdt_model_best = grid_search.best_estimator_

y_pred_best = gbdt_model_best.predict(X_test)

accuracy_best = accuracy_score(y_test, y_pred_best)

print("Best Parameters:", best_params)
print(f"Best Model Accuracy: {accuracy_best}")
```

```python

import pickle

filename = 'best_gbdt_model.pkl'

with open(filename, 'wb') as file:
    pickle.dump(gbdt_model_best, file)
```
